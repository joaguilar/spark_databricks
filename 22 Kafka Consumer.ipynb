```python
from kafka import KafkaConsumer
import json
```

# Ejemplo: Leer mensajes

Podemos entonces sacar mensajes de la cola `temperature`

```python
consumer = KafkaConsumer('temperatura',
                     bootstrap_servers=['localhost:9092'],
                     auto_offset_reset='earliest',
                     value_deserializer=lambda m: json.loads(m.decode()))
i = 0
max_messages = 100
for msg in consumer:
    if i >= max_messages:
        break
    print(msg.value)
    i += 1
```

```python

cons = KafkaConsumer('temperatura',
                     bootstrap_servers=['localhost:9092'],
                     auto_offset_reset='earliest',
                     value_deserializer=lambda m: json.loads(m.decode()))
for msg in cons:
    print(msg.value)
```
